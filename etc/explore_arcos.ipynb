```python
import re
def csvize_line(test_line):
    to_be_subbed = [0, 11, 12, 13, 14, 16, 17, 18, 54, 63, 68, 76, 80, 131, 144]
    if test_line[0] == '>':
        return 'NDC,Ingredient Key,E77-Flag (DEA Use Only),No. Ingredients,CSA Schedule,UN Schedule,Psychotropic Sched,Trade/Product Name,Package Quantity,Package Measure,Drug Code,Salt Code,Ingredient Name,Ingredient Base Weight'
    if len(test_line) < 144:
        return test_line
    reslist = []
    for x in range(0,len(to_be_subbed) -1):
#         print(to_be_subbed[x], to_be_subbed[x+1])
        res = test_line[to_be_subbed[x]:to_be_subbed[x+1]]
#         res = str(res) + ','
#         print(res + ',', len(res))
        reslist.append(str(res) + ',')
    csvd = ''.join(reslist)[:-1]
    csvd = re.sub(r'  +', '',csvd)
#     csvd = ''.join(reslist)
    
    return csvd
csvize_line(test_line)
```

```python
# create a list of strings after reading the txt file, converting each line into a comma-delimited line.
res = []
with open ('./deadiversionarcos/ndc_file-utf8.txt') as f:
    for line in f:
#         some lines include a comma which messes up the csv lineup. therefore replace all commas with a space
        line = re.sub(r',', ' ', line)
        res.append(csvize_line(line))
# res
# write the resulting list into a csv file.
with open(r'./deadiversionarcos/ndcfile.csv', 'w') as fp:
    for item in res:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('Done')
```
